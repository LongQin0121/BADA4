```python
#化路径可视化

import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize

def rosenbrock_2d(x, y):
    return (1 - x)**2 + 100 * (y - x**2)**2

def rosenbrock_vector(x):
    return rosenbrock_2d(x[0], x[1])

# 存储优化路径
path = []

def callback(x):
    path.append(x.copy())

# 从不同起点进行优化
starting_points = [[-1.2, 1.0], [0.5, -0.5], [-0.8, -0.8]]

# 创建等高线图
x = np.linspace(-2, 2, 100)
y = np.linspace(-1, 3, 100)
X, Y = np.meshgrid(x, y)
Z = rosenbrock_2d(X, Y)

plt.figure(figsize=(15, 5))

for i, start_point in enumerate(starting_points):
    path = []
    
    # 优化
    result = minimize(rosenbrock_vector, start_point, method='BFGS', 
                     callback=callback, options={'maxiter': 100})
    
    plt.subplot(1, 3, i+1)
    
    # 绘制等高线
    contour = plt.contour(X, Y, Z, levels=20, alpha=0.6)
    plt.colorbar(contour)
    
    # 绘制优化路径
    if path:
        path = np.array(path)
        plt.plot(path[:, 0], path[:, 1], 'ro-', markersize=4, 
                linewidth=2, label='优化路径')
        plt.plot(start_point[0], start_point[1], 'go', markersize=8, 
                label=f'起点{start_point}')
    
    # 标记全局最小值
    plt.plot(1, 1, 'r*', markersize=15, label='全局最小值(1,1)')
    
    plt.title(f'从{start_point}开始的优化路径')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
#牛顿法实现
import numpy as np
import matplotlib.pyplot as plt

def jacobian(x):
    """计算梯度"""
    return np.array([-400*x[0]*(x[1]-x[0]**2)-2*(1-x[0]), 
                     200*(x[1]-x[0]**2)])

def hessian(x):
    """计算Hesse矩阵"""
    return np.array([[-400*(x[1]-3*x[0]**2)+2, -400*x[0]],
                     [-400*x[0], 200]])

def newton_method(x0, max_iter=1000, tolerance=1e-5):
    """牛顿法求解"""
    x = x0.copy()
    trajectory = [x.copy()]
    
    for i in range(max_iter):
        grad = jacobian(x)
        hess = hessian(x)
        
        # 计算牛顿步长
        try:
            p = -np.dot(np.linalg.inv(hess), grad)
        except np.linalg.LinAlgError:
            print("Hesse矩阵奇异，停止迭代")
            break
            
        x_new = x + p
        
        # 检查收敛性
        if np.linalg.norm(x_new - x) < tolerance:
            print(f"在第{i+1}次迭代后收敛")
            break
            
        x = x_new
        trajectory.append(x.copy())
        
        print(f"第{i+1}次迭代: x = {x}")
    
    return x, np.array(trajectory)

# 使用牛顿法
x0 = np.array([-1.2, 1.0])
result, trajectory = newton_method(x0)
print(f"最终结果: {result}")
```

```python
import torch
import numpy as np
import matplotlib.pyplot as plt

# 定义Rosenbrock函数
def rosenbrock(x, y):
    return (1 - x)**2 + 100 * (y - x**2)**2

# 初始化变量，1.3和6.7是随意设置的初始值
x = torch.tensor(1.3, requires_grad=True)
y = torch.tensor(6.7, requires_grad=True)

# 定义SGD优化器，学习率设置为0.001
optimizer = torch.optim.Adam(params=[x, y], lr=0.001)

# 存储优化过程中的值，用于后续可视化
x_history = []
y_history = []
z_history = []

# 开始优化循环
for epoch in range(30000):
    optimizer.zero_grad()  # 清空梯度
    
    z = rosenbrock(x, y)   # 计算函数值
    z.backward()           # 反向传播计算梯度
    optimizer.step()       # 更新参数
    
    # 记录历史值
    x_history.append(x.item())
    y_history.append(y.item())
    z_history.append(z.item())     
    
    # 每1000次迭代打印一次结果
    if (epoch + 1) % 1000 == 0:
        print(f"After {epoch + 1} iterations, x = {x.item():.3f}, y = {y.item():.3f}, z = {z.item():.3f}")

print(f"\n最终结果:")
print(f"x = {x.item():.6f}")
print(f"y = {y.item():.6f}")
print(f"z = {z.item():.6f}")   # 将tensor转换为Python数值
print(f"理论最优解: x = 1.0, y = 1.0, z = 0.0")
```

```python
import torch
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt

# 定义Rosenbrock函数
def rosenbrock(x):
    """
    x是一个包含两个元素的tensor: [x1, x2]
    """
    return (1 - x[0])**2 + 100 * (x[1] - x[0]**2)**2

# 初始化参数
x = torch.tensor([1.3, 6.7], requires_grad=True)

# 定义L-BFGS优化器
optimizer = optim.LBFGS([x], 
                       lr=1.0,
                       max_iter=20,
                       history_size=10,
                       line_search_fn="strong_wolfe")

# 存储优化历史
history = []
iteration_count = 0

# 定义closure函数（L-BFGS必需）
def closure():
    global iteration_count
    optimizer.zero_grad()
    loss = rosenbrock(x)
    loss.backward()
    
    # 记录历史
    history.append(loss.item())
    iteration_count += 1
    
    if iteration_count % 5 == 0:
        print(f"Iteration {iteration_count}: x = [{x[0].item():.6f}, {x[1].item():.6f}], loss = {loss.item():.8f}")
    
    return loss

# 执行优化
print("开始L-BFGS优化...")
print(f"初始值: x = [{x[0].item():.6f}, {x[1].item():.6f}], loss = {rosenbrock(x).item():.8f}")

for i in range(50):  # 最多50次外部迭代
    optimizer.step(closure)
    
    # 检查收敛
    if len(history) > 1 and abs(history[-1] - history[-2]) < 1e-10:
        print(f"在第{i+1}次迭代后收敛")
        break

print(f"\n最终结果:")
print(f"x = [{x[0].item():.8f}, {x[1].item():.8f}]")
print(f"函数值 = {rosenbrock(x).item():.12f}")
print(f"理论最优解: x = [1.0, 1.0], 函数值 = 0.0")
print(f"总函数评估次数: {len(history)}")

```

```python
import torch
import torch.optim as optim
import matplotlib.pyplot as plt

def rosenbrock(x):
    return (1 - x[0])**2 + 100 * (x[1] - x[0]**2)**2

# SGD优化
def optimize_with_sgd():
    x_sgd = torch.tensor([1.3, 6.7], requires_grad=True)
    optimizer_sgd = optim.SGD([x_sgd], lr=0.001)
    history_sgd = []
    
    for i in range(10000):
        optimizer_sgd.zero_grad()
        loss = rosenbrock(x_sgd)
        loss.backward()
        optimizer_sgd.step()
        history_sgd.append(loss.item())
        
        if i % 1000 == 0:
            print(f"SGD Iteration {i}: loss = {loss.item():.8f}")
    
    return history_sgd, x_sgd

# L-BFGS优化
def optimize_with_lbfgs():
    x_lbfgs = torch.tensor([1.3, 6.7], requires_grad=True)
    optimizer_lbfgs = optim.LBFGS([x_lbfgs], 
                                  lr=1.0,
                                  max_iter=20,
                                  history_size=10,
                                  line_search_fn="strong_wolfe")
    history_lbfgs = []
    
    def closure():
        optimizer_lbfgs.zero_grad()
        loss = rosenbrock(x_lbfgs)
        loss.backward()
        history_lbfgs.append(loss.item())
        return loss
    
    for i in range(100):
        optimizer_lbfgs.step(closure)
        if len(history_lbfgs) > 1 and abs(history_lbfgs[-1] - history_lbfgs[-2]) < 1e-12:
            break
    
    return history_lbfgs, x_lbfgs

# 执行两种优化
print("=== SGD优化 ===")
history_sgd, x_sgd_final = optimize_with_sgd()

print("\n=== L-BFGS优化 ===")
history_lbfgs, x_lbfgs_final = optimize_with_lbfgs()

# 结果对比
print(f"\n=== 结果对比 ===")
print(f"SGD最终结果: x = [{x_sgd_final[0].item():.6f}, {x_sgd_final[1].item():.6f}], loss = {rosenbrock(x_sgd_final).item():.8f}")
print(f"L-BFGS最终结果: x = [{x_lbfgs_final[0].item():.6f}, {x_lbfgs_final[1].item():.6f}], loss = {rosenbrock(x_lbfgs_final).item():.8f}")
print(f"SGD函数评估次数: {len(history_sgd)}")
print(f"L-BFGS函数评估次数: {len(history_lbfgs)}")
```
